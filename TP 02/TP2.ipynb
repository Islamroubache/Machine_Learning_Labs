### Compute IRQ,lower and upper bounds 

```python
def get_params(df, features):
    q1 = df[features].quantile(0.25).to_numpy()
    q3 = df[features].quantile(0.75).to_numpy()
    IRQ = q3 - q1
    lower_bounds = q1 - 1.5 * IRQ
    upper_bounds = q3 + 1.5 * IRQ

    params = pd.DataFrame(
        data=np.vstack([q1,q3,IRQ,lower_bounds,upper_bounds]),
        columns=features,
        index=['q1','q3','IRQ','lower_bound','upper_bound']
    )

    return params
```

```python
params = get_params(df, ['bilirubin', 'alk_phosphate'])
params
```

### Handing outliers

```python
def handle_outliers(df, features):
    params = get_params(df, features)
    below_lower_bound = df[features] < params.loc['lower_bound']
    above_upper_bound = df[features] > params.loc['upper_bound']
    for feature in features:
        df.loc[below_lower_bound[feature], feature] = params[feature].loc['lower_bound']
        df.loc[above_upper_bound[feature], feature] = params[feature].loc['upper_bound']
    return df
```

```python
df = handle_outliers(df, ['bilirubin', 'alk_phosphate'])
```

### How the correlation matrix is really calculated

the correlation matrix is n by n matrix,where n is the number of features (including the target column), this matrix can help in detecting strong dependence between different columns, a cell in this matrix that corresponds to tow columns x and y is given by the following formula:

$$
    c_{x,y} = \frac{\sum_{i} {(x_{i} - \bar{x}) * (y_{i} - \bar{y})}}{\sqrt{\sum_{i} (x_{i} - \bar{x})^{2} * \sum_{i} (y_{i} - \bar{y})^{2}}} = \frac{cov(x,y)}{\sigma_{x} * \sigma_{y}}
$$

below is the an implementation of the algorithm that allow us to compute the correlation matrix given a data frame.

```python
def correlation(df):
    corr = np.zeros(shape=(len(df.columns), len(df.columns)),dtype=np.float64)
    for i in range(len(df.columns)):
        for j in range(len(df.columns)):
            if i < j:
                x = df[df.columns[i]]
                y = df[df.columns[j]]

                x_mean = x.mean()
                y_mean = y.mean()

                x = x - x_mean
                y = y - y_mean

                a = (x * y).sum()
                b = np.sqrt((x ** 2).sum() * (y ** 2).sum())

                corr[i][j] = a / b
                corr[j][i] = corr[i][j]
            elif i == j:
                corr[i][j] = 1
    return pd.DataFrame(corr,index=df.columns,columns=df.columns)
```
